```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
data = pd.read_csv('/data/wangxin/work/velo/RNA_mito/smartSeq2/embryo_d16/cellsnp2b/20230612_Mut_freq_SS2.csv')
```

```python
grps = data.groupby('CellType').groups
```

```python
fig, ax = plt.subplots(3, 4, figsize=(10, 8))
for ind, ct in enumerate(grps.keys()):
    # ax[ind//4][ind%4].hist(data.iloc[grps[ct]].freq)
    sns.histplot(data.iloc[grps[ct]].freq, ax=ax[ind//4][ind%4], bins=10)
    ax[ind//4][ind%4].set_title(ct)
    ax[ind//4][ind%4].set_xlim((0, 0.6))
plt.tight_layout()
```
