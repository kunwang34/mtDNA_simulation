```python
library(PATH)
library(expm)
library(ggplot2)
library(ggtree)
library(ggtreeExtra)
library(tidytree)
library(Matrix)
library(patchwork)

options(digits = 2)
```

```python
library(ComplexHeatmap)
library(grid)
library(ggnewscale)
library(viridis)
```

```python
tree <- read.tree('/home/wangkun/mtDNA_simulation/data/YFL_mtSNV_final.phy_nj.nwk')
metadata <- read.table('/home/wangkun/mtDNA_simulation/data/YFL_tree_color.txt', sep=' ', header=F, row.names=1)
tree$states <- metadata[tree$tip.label,'V2']
tree <- drop.tip(tree,tree$tip.label[is.na(tree$states)])
tree$states <- metadata[tree$tip.label,'V2']
tree$edge.length <- rep(1,length(tree$edge.length))
Pinf <- PATH_inf(tree = tree, cell_states = "states", nstates = length(unique(tree$states)))
```

```python
options(repr.plot.width=10, repr.plot.height=9)
Heatmap(Pinf$P,rocket(256),
            row_names_gp = gpar(fontsize = 8, fontface = "bold"), 
            column_names_gp = gpar(fontsize = 8, fontface = "bold"),
            cluster_rows = F,cluster_columns = F,row_names_side = 'left',column_names_side = 'top',column_names_rot = 45,
            cell_fun = function(j, i, x, y, width, height, fill) {
                    grid.text(sprintf("%.2f", as.matrix(Pinf$P)[i, j]), x, y, gp = gpar(fontsize = 15,col = "white"))
                    },
            heatmap_legend_param = list(title = "state transitions")
                   )
```
